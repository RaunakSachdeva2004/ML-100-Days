```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# We are disguising ourselves as a browser asking for data
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies', headers = headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
# to get name of the companies
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
# to get rating text
for i in soup.find_all(class_='rating_text'):
    print(i.text.strip())
```

```python
# getting reviews
len(soup.find_all('a',class_='companyCardWrapper__ActionWrapper'))
```
